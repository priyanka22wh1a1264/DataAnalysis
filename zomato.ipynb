```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=pd.read_excel("/content/zomato_rest_data.xlsx")
df
```

#Exploratory Data Analysis (EDA)

Task 1: Data Understanding & Cleaning

```python
df.info()
```

```python
#changing data type of online_order to int
df["online_order"].unique()
df["online_order"]=df["online_order"].str.replace("Yes","1")
df["online_order"]=df["online_order"].str.replace("No","0")
df["online_order"]=df["online_order"].astype(int)
df["online_order"].unique()
df.info()
```

```python
#changing data type of book_table to int
df["book_table"].unique()
df["book_table"]=df["book_table"].str.replace("Yes","1")
df["book_table"]=df["book_table"].str.replace("No","0")
df["book_table"]=df["book_table"].astype(int)
df["book_table"].unique()
df.info()
```

```python
#handling missing values and inconsistent data in rate col
df["rate"].unique()
#removing unwanted data /5
df["rate"]=df["rate"].replace("/5","",regex=True)
# changing inconsistent data like "NEW" to numeric
df["rate"]=df["rate"].replace("NEW","0")
# changing inconsistent data like "-" to numeric
df["rate"]=df["rate"].replace("-","0")

#changing the data type from object to float
df["rate"]=df["rate"].astype(float)
df["rate"].unique()
# replacing missing values with mean
mean=round(df["rate"].mean())
df["rate"]=df["rate"].fillna(mean)
df["rate"].unique()
df["rate"].isnull().sum()
```

```python
#dropping the cols which are not useful for analysis
df=df.drop(["phone_split","url","alternate_phone","address","reviews_list","menu_item","Unnamed: 0","location","phone"],axis=1)
df.info()
```

```python
#handling missing values in rest_type col
df["rest_type"].unique()
mode=df["rest_type"].mode()[0]
mode
df["rest_type"] = df["rest_type"].fillna(mode)

df["rest_type"].isnull().sum()
```

```python
#handling missing values in dish_liked col
df["dish_liked"].unique()
mode=df["dish_liked"].mode()[0]
mode
df["dish_liked"] = df["dish_liked"].fillna(mode)
df["dish_liked"].isnull().sum()
```

```python
#handling missing values in cuisines col
df["cuisines"].unique()
mode=df["cuisines"].mode()[0]
mode
df["cuisines"] = df["cuisines"].fillna(mode)
df["cuisines"].isnull().sum()
```

```python
#handling missing values in approx_cost(for two people) col
df["approx_cost(for two people)"].unique()
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].str.replace(",","")
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].str.replace("nan","0")
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].astype(float)
median= df["approx_cost(for two people)"].median()
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].fillna(median)
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].astype(int)
df["approx_cost(for two people)"].unique()
```

```python
#dropping rest_type as listed_in(type) & rest_type are same
#renaming listed_in(type) to rest_type
# df=df.drop("rest_type",axis=1)
df["rest_type"]=df["listed_in(type)"]
df=df.drop("listed_in(type)",axis=1)
df.head()
```

#visualization

1. Top 5 locations with highest number of restaurants


```python
#locations vs number of restaurants
plt.figure(figsize=(7,5))
# Create histogram
sns.histplot(x="listed_in(city)", data=df, color="red")
# Add title and labels
plt.title("Distribution of Locations")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
# Rotate x-axis labels
plt.xticks(rotation=90)
plt.show()
```

BTM, Koramangala 7th block , Koramangala 5th block, Koramangala 4th block, Koramangala 6th block these locations have highest number of restaurants

Distribution of restaurant ratings

```python
#Distribution of Restaurant Ratings
plt.figure(figsize=(7,5))
sns.histplot(df['rate'],kde=True, color="red")
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")

plt.show()
```

most of the restaurants have rating between 3 to 5

4 is the highest rating

some restaurants even have 0 as rating

no restaurant has rated 1,2 and 5

Percentage of restaurants offering online orders

```python
#Percentage of Restaurants Offering Online Orders
plt.pie(df["online_order"].value_counts(),labels=df["online_order"].unique(),autopct="%.2f%%")
plt.title("Restaurants Offering Online Orders")
plt.legend()
plt.show()
```

most of the people order online and the percentage is 58%

the percentage of people who order offline are of 41%


Distribution of average cost for two


```python
# 4)Distribution of average cost for two
# Step 1: Calculate average cost_per_two for each location
location_avg_cost = df.groupby('listed_in(city)')['approx_cost(for two people)'].mean()
# Step 2: Plot histogram of location-wise average cost
plt.figure(figsize=(8,5))
sns.histplot(location_avg_cost,color="Red",kde=True)
plt.xlabel("Average Cost for Two (Location-wise)")
plt.ylabel("Number of Locations")
plt.title("Distribution of Location-wise Average Cost for Two")
plt.show()
```

Most locations have average costs around 450 - 500 INR

Majority of locations are under 600 INR

Few locations are very high-cost
