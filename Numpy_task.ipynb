**Numpy Task**

Part 1: Creating and Reshaping Arrays

1. Create Arrays:

Create two NumPy arrays:

▪ A 1D array of 10 equally spaced values between 0 and 50.

▪ A 2D array of shape (3x3) containing random integers
between 5 and 20.

```python
import numpy as np
```

```python
a1=np.linspace(1,50,10)
a1
a2=np.random.randint(5,21,9)
a3=a2.reshape(3,3)
a3
```

2. Array Dimensions:

Print the dimensions of both arrays.

Print the shape of both arrays.


```python
print(a1.ndim,a3.ndim)
print(a1.shape,a3.shape)
```

3. Reshaping:

Reshape the 1D array into a 2D array with shape (2x5).

Print the reshaped array.

Flatten the reshaped array back to 1D and print it.

```python
a1_reshaped = a1.reshape(2,5)
print(a1_reshaped)
a1_f = a1_reshaped.flatten()
print(a1_f)
```

Part 2: Indexing and Slicing

4. Indexing:

For the 2D array created in Part 1, print the element present at the
2nd row and 3rd column.

Extract the first two rows and the last two columns of the array.

```python
print(a3)
print(a3[:2, -2:])
```

5. Slicing:

Slice the first four elements from the flattened 1D array.

Reverse the elements of the 1D array using slicing.

```python
print("\nFirst four elements of flattened 1D array:", a1_f[:4])
print("Reversed 1D array:", a1_f[::-1])
```

Part 3: Array Operations

6. Array Arithmetic:

Create another 1D array of 10 integers between 1 and 10.

Perform the following element-wise operations between the 1D

arrays created in Part 1 and Part 3:

▪ Addition

▪ Multiplication

▪ Subtract a constant value (e.g., 5) from the array.


```python
arr1d2 = np.arange(1, 11)
print("\nAddition:\n", a1 + arr1d2)
print("Multiplication:\n", a1 * arr1d2)
print("Subtract 5 from array:\n", a1 - 5)
```

Part 4: Random Number Generation and Copying

7. Random Numbers:

Create a 1D array of 5 random values between 0 and 1.

Create another 1D array of 5 random values between -1 and 1.


```python
rand1 = np.random.rand(5)
rand2 = np.random.uniform(-1, 1, 5)
print("Random values between 0 and 1:\n", rand1)
print("Random values between -1 and 1:\n", rand2)
```

Part 5: Bonus

8. Challenge Task:

Write a program that asks the user to input a number (n), and then
generates an n x n array of random integers between 1 and 100.

Find the sum of all the elements in the array and print the result.

```python
n = int(input("Enter n: "))
a = np.random.randint(1, 101, n*n)
print("1D array:\n", a)
a_2d = a.reshape(n, n)
print("Reshaped n x n array:\n", a_2d)
print("Sum of all elements:", np.sum(a))
```
