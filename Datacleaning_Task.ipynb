#Data Cleaning on Amazon Data

```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#loading the dataset
df=pd.read_csv("/content/amazon.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
#handling actual_price
df["actual_price"].unique()
df["actual_price"] = df["actual_price"].str.replace("₹", "")
df["actual_price"] = df["actual_price"].str.replace(",", "")
df["actual_price"].unique()
df["actual_price"]=df["actual_price"].astype(float)
df["actual_price"]=df["actual_price"].astype(int)
df["actual_price"].unique()
df.info()



```

```python
df["discounted_price"].unique()
df["discounted_price"] = df["discounted_price"].str.replace("₹", "")
df["discounted_price"] = df["discounted_price"].str.replace(",", "")
df["discounted_price"]=df["discounted_price"].astype(float)
df["discounted_price"]=df["discounted_price"].astype(int)
df["discounted_price"].unique()
df.info()
```

```python
#handling rating
df["rating"].unique()
df["rating"] = df["rating"].str.replace("|", "0")
df["rating"]=df["rating"].astype(float)
df["rating"].unique()
avg_rat=round(df["rating"].mean())
avg_rat
df["rating"]=df["rating"].replace(0.0,avg_rat)
df["rating"].unique()
#df
```

```python
#med=round(df["rating_count"].median())
#med
df["rating_count"] = df["rating_count"].str.replace(",", "")
df["rating_count"] = df["rating_count"].fillna(0)
#df["rating_count"] = df["rating_count"].str.replace("NaN","0")
df["rating_count"] = df["rating_count"].astype(float)
df["rating_count"] = df["rating_count"].astype(int)
# These are the rows that were missing and now replaced with 0
df[df["rating_count"] == 0]
median_value = df["rating_count"][df["rating_count"] != 0].median()  # median ignoring zeros
df["rating_count"] = df["rating_count"].replace(0, median_value)
#df["rating_count"].fillna(df["rating_count"].median(), inplace=True)
df["rating_count"]
```
